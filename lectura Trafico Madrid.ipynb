## Madrid, zonas de Madrid. Registro de Tráfico diario.

```python
# Union de datos de estaciones de tráfico registradas en Madrid.
# Zona 1 = Interior de M-30. Zona 2 = entre M-30 y M-40
# Zona 3 = Suma de todos los aforos de 60 radares con datos diarios.
# Año 2020 y 2021 (hasta donde se tienen datos)
```

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
# se importa las 18 bases de datos
```

```python
df2001 = pd.read_csv('2001DE.csv')
df2002 = pd.read_csv('2002DE.csv')
df2003 = pd.read_csv('2003DE.csv')
df2004 = pd.read_csv('2004DE.csv')
df2005 = pd.read_csv('2005DE.csv')
df2006 = pd.read_csv('2006DE.csv')
df2007 = pd.read_csv('2007DE.csv')
df2008 = pd.read_csv('2008DE.csv')
df2009 = pd.read_csv('2009DE.csv')
df2010 = pd.read_csv('2010DE.csv')
df2011 = pd.read_csv('2011DE.csv')
df2012 = pd.read_csv('2012DE.csv')
df2101 = pd.read_csv('2101DE.csv')
df2102 = pd.read_csv('2102DE.csv')
df2103 = pd.read_csv('2103DE.csv')
df2104 = pd.read_csv('2104DE.csv')
df2105 = pd.read_csv('2105DE.csv')
df2106 = pd.read_csv('2106DE.csv')
df2107 = pd.read_csv('2107DE.csv')
```

```python
# se suman la intensidad diaria de cada apunte diario (hay 4 lineas para el mismo grupo)
# una para cada 12 horas (dia y noche, total 24 horas) y otra por sentido de tráfico.
```

```python
df2001['ID']=df2001.sum(axis=1)
df2002['ID']=df2002.sum(axis=1)
df2003['ID']=df2003.sum(axis=1)
df2004['ID']=df2004.sum(axis=1)
df2005['ID']=df2005.sum(axis=1)
df2006['ID']=df2006.sum(axis=1)
df2007['ID']=df2007.sum(axis=1)
df2008['ID']=df2008.sum(axis=1)
df2009['ID']=df2009.sum(axis=1)
df2010['ID']=df2010.sum(axis=1)
df2011['ID']=df2011.sum(axis=1)
df2012['ID']=df2012.sum(axis=1)
df2101['ID']=df2101.sum(axis=1)
df2102['ID']=df2102.sum(axis=1)
df2103['ID']=df2103.sum(axis=1)
df2104['ID']=df2104.sum(axis=1)
df2105['ID']=df2105.sum(axis=1)
df2106['ID']=df2106.sum(axis=1)
df2107['ID']=df2107.sum(axis=1)
```

```python
# se imprime una base de datos para ver sus columnas, la última es la creada suma de Intensidad de
# tráfico en 12 horas.

df2107.head()
```

```python
# Agrupamos todos los datos en una sola base de datos.
dfMadrid = pd.concat([df2001, df2001, df2002, df2003, df2004, df2005, df2006, 
                      df2007, df2008, df2009, df2010, df2011, df2012, 
                      df2101, df2102, df2103, df2104, df2105, df2106, df2107])
```

```python
# Zona interior a M-30. Elijo los datos de las 40 estaciones.
```

```python
df1 = dfMadrid.loc[:, ['FDIA', 'FEST', 'ID']]
```

```python
dfMadrid_intM30ES = df1[df1.FEST.isin(['ES01', 'ES02', 'ES03', 'ES04', 'ES05', 'ES08',
                                 'ES09', 'ES12', 'ES13', 'ES14', 'ES15', 'ES16',
                                 'ES17', 'ES18', 'ES19', 'ES21', 'ES22', 'ES23',
                                 'ES24', 'ES25', 'ES26', 'ES27', 'ES28', 'ES29',
                                 'ES30', 'ES31', 'ES32', 'ES33', 'ES36', 'ES37',
                                 'ES40', 'ES41', 'ES42', 'ES44', 'ES45', 'ES48',
                                 'ES49', 'ES50', 'ES51', 'ES52'])]
```

```python
# Ya tenemos los datos de Intensidad Diaria por dia de las 40 estaciones
# en el interior de la M-30 en Madrid.
dfMadrid_intM30ES
```

```python
# solo falta agrupar los datos por día.

dfMadrid_intM30 = dfMadrid_intM30ES.groupby(['FDIA'])[['ID']].sum()
dfMadrid_intM30
```

```python
# guardamos los datos en un archivo, y examinamos antes de comenzar a trabajar en el estudio temporal.

dfMadrid_intM30.to_csv('TrafMadridintM30.csv')
```

```python
# Zona entre M-30 y M-40. Elijo los datos de las 12 estaciones.
```

```python
dfMadrid_M30M40ES = df1[df1.FEST.isin(['ES06', 'ES07', 'ES10', 'ES38', 'ES43',
                                       'ES46', 'ES47', 'ES53', 'ES54', 'ES56',
                                       'ES57', 'ES60'])]
```

```python
# Ya tenemos los datos de Intensidad Diaria por dia de las 12 estaciones
# entre la M-30 y M-40 de Madrid.
dfMadrid_M30M40ES
```

```python
# solo falta agrupar los datos por día.

dfMadrid_M30M40 = dfMadrid_M30M40ES.groupby(['FDIA'])[['ID']].sum()
dfMadrid_M30M40
```

```python
# guardamos los datos en un archivo, y examinamos antes de comenzar a trabajar en el estudio temporal.

dfMadrid_M30M40.to_csv('TrafMadridiM30M40.csv')
```
